```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

# Investigate new data

```python
new_noise_df = pd.read_csv('export_20240704_Trigger Noise_MG_0.csv')
```

```python
noise = new_noise_df['Large Angle Light Scatter']
```

```python
# filter values smaller than 2, neg values stored as 1
noise = noise[noise > 1]
```

```python
# filter values larger than 2000, mainly debris not noise
noise = noise[noise < 2000]
```

```python
noise
```

```python
def gaussian(x, amplitude, mean, stddev):
    return amplitude * np.exp(-(x - mean)**2 / (2 * stddev**2))
```

```python
bins_noise = noise.max() - noise.min()
bins, bin_edges = np.histogram(noise, bins=int(bins_noise))
```

```python
bins_noise
```

```python
fig, axs = plt.subplots(1, 1, figsize=(10, 10))

axs.step(bin_edges[:-1], bins,  linewidth=4, label='Data')

# plt.xlim(2, 2000)
plt.xlabel('LALS (a.u.)')
plt.ylabel('Counts')
```

```python
np.max(bins)
```

```python
initial_amplitude = np.max(bins)
initial_std = np.std(noise)
initial_mean = np.mean(noise)

initial_guess = [initial_amplitude, initial_mean, initial_std]  # Initial values for amplitude, mean, and standard deviation
params, covariance = curve_fit(gaussian, bin_edges[:-1], bins, p0=initial_guess)

# Extract the fitted parameters
amplitude_fit, mean_fit, stddev_fit = params
amplitude_fit, mean_fit, stddev_fit

y_fit = gaussian(bin_edges[:-1], amplitude_fit, mean_fit, stddev_fit)
```

```python
fig, axs = plt.subplots(1, 1, figsize=(10, 10))

axs.step(bin_edges[:-1], bins,  linewidth=4, label='Data')
axs.step(bin_edges[:-1], y_fit,  linewidth=4, label='Fit')

# plt.xlim(-1, 6)
plt.xlabel('PE (a.u.)')
plt.ylabel('Counts')
```

# Investigate data from prev summer school

```python
old_noise_df = pd.read_csv('./Day 2a Trigger threshold_Apogee/export_MQ_300V_LALS_triggerFITC_0.csv')
```

```python
noise = old_noise_df['Large Angle Light Scatter']
```

```python
# filter values smaller than 2, neg values stored as 1
noise = noise[noise > 1]
```

```python
print(np.percentile(noise, 99.95))
```

```python
# filter values larger than 2000, mainly debris not noise
print(np.percentile(noise, 99.95))
noise = noise[noise < np.percentile(noise, 99.95)]
```

```python
np.max(noise)
```

```python
bins_noise = noise.max() - noise.min()
bins, bin_edges = np.histogram(noise, bins=int(bins_noise))
```

```python
fig, axs = plt.subplots(1, 1, figsize=(10, 10))

axs.step(bin_edges[:-1], bins,  linewidth=4, label='Data')

# plt.xlim(2, 2000)
plt.xlabel('LALS (a.u.)')
plt.ylabel('Counts')
```

```python
def fit(noise, factor=1):
    bins_noise = noise.max() - noise.min()
    bins, bin_edges = np.histogram(noise, bins=int(bins_noise/factor))
    # print(bins)
    
    initial_amplitude = np.max(bins)
    initial_std = np.std(noise)
    initial_mean = np.mean(noise)
    
    initial_guess = [initial_amplitude, initial_mean, initial_std]  # Initial values for amplitude, mean, and standard deviation
    params, covariance, infodict, _, _ = curve_fit(gaussian, bin_edges[:-1], bins, p0=initial_guess, full_output=True)
    
    # Extract the fitted parameters
    amplitude_fit, mean_fit, stddev_fit = params
    amplitude_fit, mean_fit, stddev_fit

    residuals = infodict['fvec']
    
    return bins, bin_edges, amplitude_fit, mean_fit, stddev_fit, residuals
```

```python
bins, bin_edges, amplitude_fit, mean_fit, stddev_fit, residuals = fit(noise, factor=10)
y_fit = gaussian(bin_edges[:-1], amplitude_fit, mean_fit, stddev_fit)

fig, axs = plt.subplots(1, 1, figsize=(10, 10))

axs.step(bin_edges[:-1], bins,  linewidth=4, label='Data')
axs.step(bin_edges[:-1], y_fit,  linewidth=4, label='Fit')

# plt.xlim(1, 2000)
plt.xlabel('PE (a.u.)')
plt.ylabel('Counts')
print(np.sum(residuals**2))
lod= mean_fit + (4*stddev_fit)
lod/(2**6)
```
